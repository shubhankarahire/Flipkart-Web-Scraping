```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import numpy as nP
import pandas as pd
```

```python
# Make a request to the Flipkart website and get the page content
page = requests.get("https://www.flipkart.com/search?q=i+phone+13&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=i+phone+13%7CMobiles&requestId=742671e4-2797-4952-8095-c49aab1b3534&as-searchtext=i%20phone%2013")
page
```

```python
# Display the page object
page.content
```

```python
# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(page.content,"html")

# Display the BeautifulSoup object
soup
```

```python
# Find the element with id "container"
id_class = soup.find(id="container")
print(id_class)
```

```python
# Find all elements with class "_2kHMtA" inside the "container" element
forcast_item = id_class.find_all(class_ = "_2kHMtA")
print(forcast_item)
```

```python
# Display the prettified HTML of the first element in forcast_item
f1 = forcast_item[0]
print(f1.prettify())
```

```python
# Find all elements with class "_4rR01T"
name = soup.find_all(class_ = "_4rR01T")
name
```

```python
# Extract text from each element and store in name1 list
name1 = []
for i in name:
    text = i.get_text()
    name1.append(text)

# Display the extracted names
name1
```

```python
# Find the price element and extract the text
price = soup.find(class_ = "_30jeq3 _1_WHN1").get_text()
price
```

```python
# Find all elements with class "_30jeq3 _1_WHN1"
price = soup.find_all(class_ = "_30jeq3 _1_WHN1")
price1 = []

# Extract text from each element and store in price1 list
for i in price:
    text = i.get_text()
    price1.append(text)

# Display the extracted prices
price1
```

```python
# Find the element with class "fMghEO" and extract the text
discription = soup.find(class_= "fMghEO").get_text()
discription
```

```python
# Find the element with class "gUuXy-" and extract the text, then split and convert to float
rating = soup.find(class_ = "gUuXy-").get_text()
rating = rating.split(" ")[0]
rating
```

```python
# Find all elements with class "gUuXy-"
rating = soup.find_all(class_ = "gUuXy-")
rating1 = []

# Extract and process text from each element and store in rating1 list
for i in rating:
    text = i.get_text().split(" ")[0]
    text = text.split(",")[0]
    text = float(text)
    rating1.append(text)
    
# Display the extracted ratings
rating1
```

```python
# Find all elements with class "fMghEO"
discription = soup.find_all(class_= "fMghEO")
discription1 = []

# Extract text from each element and store in discription1 list
for i in discription:
    text = i.get_text()
    discription1.append(text)

# Display the extracted descriptions
discription1
```

```python
# Create a dictionary with the extracted data
data = {
    "Mobile Name": name1,
    "discription": discription1,
    "rating": rating1,
    "Price": price1,
}

# Create a DataFrame from the dictionary

df_data = pd.DataFrame(data)
#df_data[["Mobile Name","discription","Price"]] to change column place
```

```python
# Save the DataFrame to a CSV file
df_data
```
